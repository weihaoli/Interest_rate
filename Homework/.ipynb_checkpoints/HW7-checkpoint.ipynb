```python
def wiener_process(n, t, seed=12345):
    dx = sobol_seq.i4_sobol_generate(1, n, seed).flatten()
    dx = norm.ppf(dx)
    C = np.array([[ (1+min(i, j))*float(t)/n for j in range(n)] for i in range(n)])
    lam, E = np.linalg.eig(C)
    w = np.array([sum([ np.sqrt(lam[i])*E[j, i]*dx[i] for i in range(n)]) for j in range(n)])
    dw = np.diff(w)
    dw = np.concatenate([[0],dw])
    return dw, w

class Libor_Market:
    def __init__(self, sigma, rho, spread, horizon, maturity, L0, dt, frozen=False, period=.25):
        self.period = period
        self.sigma = sigma
        self.rho = rho
        self.spread = spread
        self.horizon = horizon
        self.maturity = maturity
        self.N = int(maturity/self.period+1)
        self.dt = dt
        self.M = int(horizon/dt)
        self.paths = np.zeros([self.M+1, self.N])
        self.paths[0] = L0
        self.frozen = frozen
        
    def delta(self, i, t):
        res = 0
        if not self.frozen:
            for j in range(i+1, self.N):
                res -= self.rho*self.period*self.sigma/(1+self.period*self.paths[t, j])
        else:
            for j in range(2, self.N):
                res -= self.rho*self.period*self.sigma/(1+self.period*self.paths[0, j])
        return res
        
    def simulate(self, seed=134):
        BM, _ = wiener_process(self.M, self.horizon, seed)
        for i in range(1, self.M+1):
            for j in range(self.N):
                if j != self.N-1:
                    self.paths[i, j] = self.paths[i-1, j] + self.sigma*self.delta(j, i-1)*self.dt + self.sigma*BM[i-1]
                else:
                    self.paths[i, j] = self.paths[i-1, j] + self.sigma*BM[i-1]
                self.paths[i, j] = max(self.paths[i, j], 0.)
        return self.paths[-1]
```

```python
@jit
def Swap_Float(spread, L0, period=.25):
    res = 0
    for i in range(len(L0)-1):
        res += L0[i]*np.exp(-(L0[i+1]-spread)*(i+1)*period)
    return res

@jit
def Swap_Fix(spread, L0, K, period=.25):
    res = 0
    for i in range(len(L0)-1):
        res += K*np.exp(-(L0[i+1]-spread)*(i+1)*period)
    return res

@jit
def Swap_Receive(spread, L0, K, period=.25):
    res = 0
#     test = []
    for i in range(len(L0)-1):
        res -= period*L0[i]*np.exp(-(L0[i+1]-spread)*(i+1)*period)
        
    for i in range(0, len(L0)-1, 2):
        res += 2*period*K*np.exp(-(L0[i+2]-spread)*(i+2)*period)
    return res
```

```python
class Swaption:
    def __init__(self, discount, sigma, rho, spread, horizon, maturity, L0, dt, K, N = 2000, period=.25):
        self.LMM = Libor_Market(sigma, rho, spread, horizon, maturity, L0, dt, period)
        self.horizon = horizon
        self.discount = discount
        self.K = K
        self.N = N
        self.spread = spread
        
    def price(self):
        res = 0.
        for i in range(self.N):
            temp = self.LMM.simulate(np.random.randint(9999))
            res += max(Swap_Receive(self.spread, temp, self.K), 0.)
#             print(res)
        return np.exp(-self.discount*self.horizon)*res/self.N
```
